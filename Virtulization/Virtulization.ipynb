**Load Data**

```python
import pandas as pd
#import data and generate the description
pd.set_option('display.max_columns',999)
df_07_11 = pd.read_csv('https://resources.lendingclub.com/LoanStats3a.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_12_13 = pd.read_csv('https://resources.lendingclub.com/LoanStats3b.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_14 = pd.read_csv('https://resources.lendingclub.com/LoanStats3c.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_15 = pd.read_csv('https://resources.lendingclub.com/LoanStats3d.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_16Q1 = pd.read_csv('https://resources.lendingclub.com/LoanStats_2016Q1.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_16Q2 = pd.read_csv('https://resources.lendingclub.com/LoanStats_2016Q2.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_16Q3 = pd.read_csv('https://resources.lendingclub.com/LoanStats_2016Q3.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_16Q4 = pd.read_csv('https://resources.lendingclub.com/LoanStats_2016Q4.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_17Q1 = pd.read_csv('https://resources.lendingclub.com/LoanStats_2017Q1.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_17Q2 = pd.read_csv('https://resources.lendingclub.com/LoanStats_2017Q2.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_17Q3 = pd.read_csv('https://resources.lendingclub.com/LoanStats_2017Q3.csv.zip', compression='zip', skiprows=[0],low_memory=False)
df_17Q4 = pd.read_csv('https://resources.lendingclub.com/LoanStats_2017Q4.csv.zip', compression='zip', skiprows=[0],low_memory=False)

df_all=pd.concat([df_07_11,df_12_13,df_14,df_15,df_16Q1,df_16Q2,df_16Q3,df_16Q4,df_17Q1,df_17Q2,df_17Q3,df_17Q4])

df_all.describe()
```

**Only keep loans with valid loan_status**

```python
df_all=df_all.dropna(subset=['loan_status'])
```

**Loan Status Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(y="loan_status", data=df_all)
plt.title('Loan Status')
plt.xlabel('Count')
plt.ylabel('Status')
```

**There is only a few "default", if we define bad loan using charge_off and past due, Let's see the distribution.**

```python
df_all['status']=0
df_all['status'][df_all['loan_status'].
                 str.contains("Charged Off|Does not meet the credit policy. Status:Charged Off|Default")]=1
```

```python
loan_st=df_all['status'].value_counts()
```

```python
labels ="Good Loans", "Bad Loans"
plt.suptitle('Good Loan/Bad Loan distribution')
loan_st.plot.pie(explode=[0,0.25], autopct='%1.2f%%', shadow=True, labels=labels)
```
